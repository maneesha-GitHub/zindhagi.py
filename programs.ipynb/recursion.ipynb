```python
#sum of natural numbers in recursion
def sum(n):
    if n==0:
        return 0
    return sum(n-1)+n
sum(10)
```

```python
def strong(n, i=0, s=0):
    if i == len(str(n)):
        return "strong" if n == s else "not strong"
    else:
        d = int(str(n)[i])
        fact = 1
        for j in range(1, d + 1):
            fact *= j
        s += fact
        return strong(n, i + 1, s)

n = int(input("Enter a number: "))
result = strong(n)
print(result)
```
