```python
#single inheritance
class parent:
    def fun1(self):
        print("this is parent class")

class child(parent):
    def fun2(self):
        print("this is child class")

ob=child()
ob.fun1()
```

```python
#multi inheritance
class parent:
    def fun1(self):
        print("parent class")

class child(parent):
    def fun2(self):
        print("child class")

class grand_child(child):
    def fun3(self):
        print("grand child")

obj=grand_child()
obj.fun1()
obj.fun2()
obj.fun3()
print("_"*19)
obj1=child()
obj1.fun1()
obj1.fun2()
#obj1.fun3()
print("_"*20)
ob2=parent()
ob2.fun1()
```

```python
#multiple inheritance
class parent1:
    def fun1(self):
        print("x properties")

class parent2:
    def fun2(self):
        print("y properties")
    
class child(parent1 ,parent2):
    def fun3(self):
        print(" x + y properties")

obj=child()
obj.fun2()

```

```python
class Profile:
    def __init__(self, name, age, gender, occupation):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupation = occupation
        self.matches = []  # List to store potential matches

    def view_profile(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Occupation: {self.occupation}")

    def add_match(self, profile):
        self.matches.append(profile)
        print(f"{profile.name} added to your matches!")

# Creating instances of Profile class
profile1 = Profile("Alice", 28, "Female", "Engineer")
profile2 = Profile("Bob", 30, "Male", "Doctor")

# Viewing profiles
profile1.view_profile()
profile2.view_profile()

# Adding matches
profile1.add_match(profile2)
```
